```python
from modules.backtester import long_only_backtester
from modules.data_fetcher import download_historical_data

from ta.trend import ema_indicator
```

```python
symbol = 'BTC-USDT'

df_BTC = download_historical_data(symbol,'1hour')
df_BTC["Return"] = df_BTC.Close.pct_change()
df_BTC.dropna(inplace=True)
print(df_BTC.shape)
df_BTC.head()
```

```python
df_BTC['EMA20'] = ema_indicator(df_BTC.Close,20)
df_BTC['EMA60'] = ema_indicator(df_BTC.Close,60)
df_BTC.dropna(inplace=True)

df_BTC.head()
```

```python
def buy_func(row, prev_row)->bool:
    return True if row.EMA20 > row.EMA60 else False

def sell_func(row, prev_row, trading_days)->bool:
    return True if row.EMA20 < row.EMA60 else False

long_only_backtester(df_BTC,buy_func,sell_func)
```
